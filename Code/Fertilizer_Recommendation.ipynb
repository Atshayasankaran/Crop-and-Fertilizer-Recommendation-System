 **Fertilizer Recommendation system - NPK value Prediction**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data=pd.read_csv('/content/drive/MyDrive/agriculture dataset/Crop_recommendation.csv')
data.head()
```

```python
dataMapping= {
      "apple":1,
      "banana":2,
      "blackgram":3,
      "chickpea":4,
      "coconut":5,
      "coffee":6,
      "cotton":7,
      "grapes":8,
      "jute":9,
      "kidneybeans":10,
      "lentil":11,
      "maize":12,
      "mango":13,
      "mothbeans":14,
      "mungbean":15,
      "muskmelon":16,
      "orange":17,
      "papaya":18 ,
      "pigeonpeas":19,
      "pomegranate":20,
      "rice":21,
      "watermelon":22
}
```

```python
data['label']=data['label'].map(dataMapping)
data
```

```python
npk_features = data.iloc[:,3:8].values
npk_labels = data.iloc[:, 0:3].values
print(npk_features)
```

```python
from sklearn.model_selection import train_test_split
ratio_train = 0.8
ratio_val = 0.1
ratio_test = 0.1
npk_x_remaining, npk_x_test, npk_y_remaining, npk_y_test = train_test_split(npk_features, npk_labels, test_size=ratio_test,random_state=10)
ratio_remaining = 1 - ratio_test
ratio_val_adjusted = ratio_val / ratio_remaining
npk_x_train, npk_x_val, npk_y_train, npk_y_val = train_test_split(npk_x_remaining, npk_y_remaining, test_size=ratio_val_adjusted,random_state=10)
print(npk_x_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
npk_x_train = scaler.fit_transform(npk_x_train)
npk_x_test = scaler.fit_transform(npk_x_test)
npk_x_val = scaler.fit_transform(npk_x_val)
print(npk_x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
knn=KNeighborsClassifier(n_neighbors=3)
multi_target = MultiOutputClassifier(knn)
multi_target.fit(npk_x_train,npk_y_train)
```

```python
npk_predictions = multi_target.predict(npk_x_test)
npk_predictions
```

```python
print(npk_y_test)
```

```python
npk_input = [[20., 82., 6., 202.,21.]]
npk_input = scaler.fit_transform(npk_input)

npk_output= multi_target.predict(npk_input)
npk_output
```

**Fertilizer recommendation from predicted npk value**

```python
data=pd.read_csv('/content/drive/MyDrive/agriculture dataset/fertilizer.csv',encoding='windows-1254')
data.head()
```

```python
fer_features = data.iloc[:,0:3].values
fer_labels = data.iloc[:,3].values
print(fer_features)
```

```python
from sklearn.model_selection import train_test_split
fer_x_train=fer_features
fer_y_train=fer_labels
fer_x_test=fer_features
fer_y_test=fer_labels
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
fer_x_train = scaler.fit_transform(fer_x_train)
fer_x_test = scaler.fit_transform(fer_x_test)
print(fer_x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(fer_x_train, fer_y_train)
```

```python
predictions = knn.predict(fer_x_test)
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(fer_y_test,predictions))
print(classification_report(fer_y_test,predictions))
```
